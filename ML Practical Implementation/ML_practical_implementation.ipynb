## K-Means Clustering

```python
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
```

```python
X,y = make_blobs(n_samples=1000,centers=3,n_features=2,random_state=23)
```

```python
plt.scatter(X[:,0],X[:,1])
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.33, random_state=42)
```

```python
wcss=[]
for i in range(1,11):
    KMean = KMeans(n_clusters=i,init='k-means++')
    KMean.fit(X_train) 
    wcss.append(KMean.inertia_)
```

```python
wcss
```

```python
plt.plot(range(1,11),wcss)
plt.xticks(range(1,11))
plt.show()
```

```python
KMean = KMeans(n_clusters=3,init='k-means++')
```

```python
y_label = KMean.fit_predict(X_train)
```

```python
plt.scatter(X_train[:,0],X_train[:,1], c= y_label)
plt.title('KMeans Clustering')
```

## Hierarchical Clustering

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
X = pd.DataFrame(iris.data,columns=iris.feature_names)
y = iris.target
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.33, random_state=42)
```

```python
X_train
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaled_X_train = scaler.fit_transform(X_train)
```

```python
scaled_X_train
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
pca_X_train = pca.fit_transform(scaled_X_train)
```

```python
pca_X_train.shape
```

```python
plt.scatter(pca_X_train[:,0],pca_X_train[:,1])
plt.title('Hierarchical Clustering')
plt.show()
```

```python
import scipy.cluster.hierarchy as sc
```

```python
plt.figure(figsize=(20,10))
sc.dendrogram(sc.linkage(pca_X_train,method='ward'))
plt.title('Dendograms')
plt.xlabel('Sample Index')
plt.ylabel('Eucledian Distance')
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
ag =AgglomerativeClustering(n_clusters=2)
```

```python
ag.fit(pca_X_train)
```

```python
plt.scatter(pca_X_train[:,0], pca_X_train[:,1], c=ag.labels_)
plt.title('Hierarchical Clustering')
plt.show()
```

```python
silhouette_score(pca_X_train,ag.labels_)
```
